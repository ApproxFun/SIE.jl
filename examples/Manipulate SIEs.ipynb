```julia
#Requires Gadfly
using ApproxFun, SIE, Interact
```

# Internal mode I crack in an infinite strip

For the crack opening displacement $\Delta v$:
$$\frac{P.V.}{\pi}\int_{-1}^{+1}\frac{\Delta v(y)}{(y-x)^2}{\rm d}y + \frac{1}{\pi}\int_{-1}^{+1}K(x,y)\Delta v(y){\rm d}y = p(x), \qquad \Delta v(\pm1)=0,$$
where:
$$K(x,y) = -\frac{1}{(x+y+2\epsilon)^2}+\frac{12(x+\epsilon)}{(x+y+2\epsilon)^3}-\frac{12(x+\epsilon)^2}{(x+y+2\epsilon)^4},\qquad p(x) = -1.$$

```julia
x = Fun(identity)
w = 1/sqrt(1-x^2)
d = domain(x)
d2 = d^2
B = dirichlet(d)
H2 = Hilbert(d,2)
S = Σ(d)
p = -Fun(one)

@manipulate for ϵ=1.05:0.01:1.5
    K = LowRankFun((x,y)->-1./(x+y+2ϵ).^2+12(x+ϵ)./(x+y+2ϵ).^3-12(x+ϵ).^2./(x+y+2ϵ).^4,d2)
    L = H2[w] + S[K*(w/π)]
    uSIE = [B,L]\[zeros(2),p]
    ApproxFun.plot(uSIE*w;axis=[0.0,1.5])
end
```

# Mode III crack problem in nonhomogeneous medium

For the crack opening displacement $\Delta v$:
$$G(x)\frac{P.V.}{\pi}\int_{-1}^{+1}K(x,y)\Delta v(y){\rm d}y = p(x), \qquad \Delta v(\pm1)=0,$$
where:
$$K(x,y) = \frac{\beta e^{\beta(y-x)/2}}{2}\frac{K_1(\beta|y-x|)}{y-x},\qquad G(x) = e^{\beta x},\qquad p(x) = -1.$$

```julia
include("Fractureaux.jl")

x = Fun(identity)
w = 1/sqrt(1-x^2)
d = domain(x)
d2 = d^2
B = dirichlet(d)
H0 = Hilbert(d,0)
H2 = Hilbert(d,2)
S = Σ(d)
p = -Fun(one)

@manipulate for β = -2.0:0.1:-0.5
    G = exp(β*x)
    K = LowRankFun((x,y)->exp(β/2.*(y-x)).*β^2./2.^2.*(FK1(β.*(y-x)./2).*(log(abs(β)/4) + γ) - GK1(β.*(y-x)./2)./2),d2)
    K0 = LowRankFun((x,y)->exp(β/2.*(y-x)).*β^2./2.^2.*FK1(β.*(y-x)./2),d2)
    K2 = LowRankFun((x,y)->exp(β/2.*(y-x)),d2)
    L = H2[K2*w] + H0[K0*w] + S[K*(w/π)]
    uSIE = [B,L]\[zeros(2),p/G]
    ApproxFun.plot(uSIE*w/G[1],axis=[0.0,12.0])
end
```

# Gradient elasticity applied to mode III cracks

For the crack opening displacement $\Delta v$:
$$\frac{P.V.}{\pi}\int_{-1}^{+1}\left(\frac{-6\epsilon^2}{(y-x)^4} + \frac{1}{(y-x)^2}\right)\Delta v(y){\rm d}y = p(x), \qquad \Delta v(\pm1)=\Delta v'(\pm1)=0,\qquad {\rm where} \qquad p(x) = -1.$$

```julia
x = Fun(identity)
w = 1/sqrt(1-x^2)
d = domain(x)
d2 = d^2
B = [dirichlet(d),neumann(d)]
H2 = Hilbert(d,2)
H4 = Hilbert(d,4)
p = -Fun(one)
@manipulate for ϵ = 0.0005:0.0005:0.4
    L = -6ϵ^2*H4[w] + H2[w]
    uSIE = [B,L]\[zeros(4),p]
    ApproxFun.plot(uSIE*w;axis=[0.0,1.0])
end
```
